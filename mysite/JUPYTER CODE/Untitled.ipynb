```python
# storing and anaysis
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("./valeursfoncieres-2020.txt", sep="|")
data.head()
```

```python
data.drop(['Code service CH','Reference document','1 Articles CGI','2 Articles CGI','3 Articles CGI','4 Articles CGI','5 Articles CGI','No disposition'], axis=1, inplace=True)
data.drop(['No voie','B/T/Q'],axis=1, inplace=True)
data.drop(['Prefixe de section','Section','No plan','No Volume','1er lot','Surface Carrez du 1er lot','2eme lot','Surface Carrez du 2eme lot','3eme lot','Surface Carrez du 3eme lot','4eme lot','Surface Carrez du 4eme lot','5eme lot','Surface Carrez du 5eme lot','Nombre de lots','Code type local','Identifiant local','Nature culture'],axis=1, inplace=True)
data
```

```python
#data = data.dropna()
data["Valeur fonciere"] = data["Valeur fonciere"].str.replace(',','.')
data
```

```python
data["Valeur fonciere"] = pd.to_numeric(data["Valeur fonciere"])
data["Surface reelle bati"] = pd.to_numeric(data["Surface reelle bati"])
```

```python
# On calcul le prix au metre carré de chaque commune
data['Prix m^2'] = data['Valeur fonciere'] / data['Surface reelle bati']
temp = data.groupby(['Code departement'])['Prix m^2']
temp.style.background_gradient(cmap='Blues')
```
